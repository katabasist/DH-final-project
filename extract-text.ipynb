```python
import os

pdf_folder = "episode-scripts"
output_folder = "episode-text"
```

```python
import pdfplumber

for pdf_file in os.listdir(pdf_folder):
    if pdf_file.endswith(".pdf"):
        # Generate output filename (replace .pdf with .txt)
        txt_file = pdf_file.replace(".pdf", ".txt")
        output_path = os.path.join(output_folder, txt_file)

        # Extract text from PDF
        full_text = []
        with pdfplumber.open(os.path.join(pdf_folder, pdf_file)) as pdf:
            for page in pdf.pages:
                text = page.extract_text()
                if text:
                    full_text.append(text)

        # Save to .txt file
        with open(output_path, "w", encoding="utf-8") as f:
            f.write("\n".join(full_text))
        print(f"Extracted: {pdf_file} → {txt_file}")
```
